[문제 3]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요. 

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
numbers = []

while True:
    nums = input('숫자를 입력해주세요. (종료 : done): ')

    if nums == 'done':
        break
    try:
        nums = int(nums)
        numbers.append(nums)
        print(f'숫자 {nums} 가 추가되었습니다.')
    except ValueError:
        print(f'잘못된 값이 있습니다.{nums}, 숫자를 입력해주세요.')

print('최종 리스트 :', numbers)
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        print(f"함수 실행 시간: {time.perf_counter() - start:.2f}초")
        return result
    return wrapper

@timer_decorator
def time_test():
    time.sleep(1)
    print("작업 완료!")

time_test()
````

[문제 8]

여러분이 자주 사용하는 수학적 연산<br>
(예: 삼각형 넓이, 원의 넓이, 직육면체의 넓이)을 수행하는 함수들을 포함하는 mymath.py라는 이름의 모듈을 만드세요. 

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
# mymath.py
pi = 3.141592

#삼각형의 넓이
def triangle(base, height):
    return base * height * 0.5

# 원의 넓이
def circle(radius):
    return pi * radius ** 2

# 정사각형의 넓이
def square(length, width, height):
    return 2 * (length * width + width * height + height * length)
````

````python
# math
import mymath as math

base, height = map(int, input('삼각형의 밑변과 높이를 입력해주세요. :').split(','))
triangle_area = math.triangle(base, height)

radius = int(input('원의 반지름을 입력해주세요. :'))
circle_area = math.circle(radius)

length, width, height = map(int, input('직육면체의 가로, 세로, 높이를 입력해주세요. :').split(','))
rect_area = math.square(length, width, height)

print(f'삼각형의 넓이 : {triangle_area}')
print(f'원의 넓이 : {circle_area}')
print(f'직육면체의 넓이 : {rect_area}')
````
