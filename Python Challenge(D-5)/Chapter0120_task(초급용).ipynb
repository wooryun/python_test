설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
    # 1) bread_type이 self.stock에 존재하는지 확인하세요.
    #    - 없다면 "없는 메뉴입니다." 메시지를 출력하고 return 처리.

    # 2) 재고가 부족하면 "재고가 부족합니다." 메시지를 출력하고 return 처리.

    # 3) 재고가 충분하면:
    #    - 재고에서 주문량을 빼세요.
    #    - 판매량을 주문량만큼 더하세요.
    #    - 주문 완료 메시지를 출력하세요. 예: "주문이 완료되었습니다. {bread_type} {bread_count}개"
class BungeoppangShop:
    def process_order(self, bread_type, bread_count):
        if bread_type not in self.stock:
            print("없는 메뉴입니다.")
            return
        if self.stock[bread_type] < bread_count:
            print("재고가 부족합니다.")
            return
        self.stock[bread_type] -= bread_count
        self.sales[bread_type] += bread_count
        print(f"주문이 완료되었습니다. {bread_type} {bread_count}개")
```

```python
    # 1) BungeoppangShop 객체를 생성하세요.
    
    # 2) 프로그램 시작 메시지를 출력하세요.
    
    # 3) while 루프를 만들어 프로그램을 계속 실행합니다.
    #    - 사용자에게 "주문", "관리자", "종료" 중 하나를 입력받습니다.
    #    - "주문" 모드:
    #         - 주문할 붕어빵 종류와 개수를 입력받아 process_order 호출.
    #         - 주문 후 check_stock 호출.
    #    - "관리자" 모드:
    #         - admin_mode 호출.
    #    - "종료"를 입력하면 루프를 종료하세요.
    #    - 잘못된 입력이라면 "잘못된 입력입니다."를 출력하세요.
    
    # 4) 프로그램 종료 시 calculate_total_sales 호출.
    # 5) 종료 메시지를 출력하세요.
def main():
    shop = BungeoppangShop()
    print("붕어빵 가게에 오신 것을 환영합니다.")

    while True:
        mode = input("주문, 관리자, 종료 중 하나를 입력해주세요.")

        if mode == "주문":
            bread_type = input("주문할 붕어빵 종류를 입력해주세요. :")
            bread_count = int(input("주문할 붕어빵 개수를 입력해주세요.:"))
            shop.process_order(bread_type, bread_count)
            shop.check_stock()

        elif mode == "관리자":
            shop.admin_mode()
            shop.check_stock()

        elif mode == "종료":
            print("프로그램을 종료합니다. 이용해 주셔서 감사합니다.")
            break

        else:
            print("잘못된 입력입니다.")
```

__[과제 7: 최종 BungeoppangShop 클래스]__

함수로 리팩토링한 "붕어빵 자동 판매 시스템" 작성 후 실행합니다. 

```python
class BungeoppangShop:
    def __init__(self):
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
            }
        self.prices = {
            "팥붕어빵" : 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
            }
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
            }

    def check_stock(self):
        for bread, count in self.stock.items():
            print(f"{bread}: {count}개")

    def process_order(self, bread_type, bread_count):
        if bread_type not in self.stock:
            print("없는 메뉴입니다.")
            return
        if self.stock[bread_type] < bread_count:
            print("재고가 부족합니다.")
            return
        self.stock[bread_type] -= bread_count
        self.sales[bread_type] += bread_count
        print(f"주문이 완료되었습니다. {bread_type} {bread_count}개")

    def admin_mode(self):
        print("관리자 모드로 진입했습니다.")
        bread_type = input("추가할 붕어빵 종류를 입력해주세요 :")
        additional_stock = int(input("추가할 재고량을 입력해주세요 :"))
        if bread_type not in self.stock:
            print("없는 메뉴입니다.")
        else:
            self.stock[bread_type] += additional_stock
            print(f'{bread_type} 재고가 {additional_stock}개 추가되었습니다.')

    def calculate_total_sales(self):
        total_sales = 0
        for bread, count in self.sales.items():
            total_sales += count * self.prices[bread]
        print(f"총 매출: {total_sales}원")

def main():
    shop = BungeoppangShop()
    print("붕어빵 가게에 오신 것을 환영합니다.")

    while True:
        mode = input("주문, 관리자, 종료 중 하나를 입력해주세요.")

        if mode == "주문":
            bread_type = input("주문할 붕어빵 종류를 입력해주세요. (팥붕어빵, 슈크림붕어빵, 초코붕어빵):")
            bread_count = int(input("주문할 붕어빵 개수를 입력해주세요.:"))
            shop.process_order(bread_type, bread_count)
            shop.check_stock()

        elif mode == "관리자":
            shop.admin_mode()
            shop.check_stock()

        elif mode == "종료":
            print("프로그램을 종료합니다. 이용해 주셔서 감사합니다.")
            break

        else:
            print("잘못된 입력입니다.")

    shop.calculate_total_sales()

if __name__ == "__main__":
    main()
```
